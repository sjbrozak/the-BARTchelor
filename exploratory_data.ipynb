```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load data
full_df = pd.read_csv('bachelor_full.csv')
full_df.head
```

```python
full_df.plot.hist(column = 'Age',figsize = (6,3));
```

```python
# count the values to plot pie chart
r0 = full_df['Race_0'].value_counts()[1]   # White contestants
r1 = full_df['Race_1'].value_counts()[1]   # Black contestants
r2 = full_df['Race_2'].value_counts()[1]   # Asian contestants
r3 = full_df['Race_3'].value_counts()[1]   # Hispanic contestants
r4 = full_df['Race_4'].value_counts()[1]   # Middle Eastern contestants

plt.pie([r0,r1,r4,r2,r3],labels=['White','Black','Middle Eastern','Asian','Hispanic']);
```

```python
# view distribution of overall place by 1-on-1 week
one_on_one_df = full_df.loc[full_df['1-on-1 week'] != 0,['1-on-1 week','Place']]
one_on_one_df.boxplot(by='1-on-1 week')
```

```python
full_df.boxplot(column = '1-on-1 week',by='Place')
```
